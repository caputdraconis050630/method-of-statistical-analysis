```python
# 1
import matplotlib.pyplot as plt
plt.style.use('ggplot')

customers = ['A', 'D', 'G', 'J', 'M']
customers_index = range(len(customers)) # range(5) => [0, 1, 2, 3, 4]
sale_amounts = [1270, 900, 2010, 1110, 2320]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar(customers_index, sale_amounts, align='center', color='blue')
plt.xticks(customers_index, customers, fontsize='small')

plt.title('Sale Amount Per Customer')
plt.xlabel('Customer Name')
plt.ylabel('Sale Amount')

plt.show()
```

```python
# 2
import numpy as np

mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma * np.random.randn(10000)
x2 = mu2 + sigma * np.random.randn(10000)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
n, bins, patches = ax.hist(x1, bins=50, color='green')
n, bins, patches = ax.hist(x2, bins=50, color='orange')

plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
fig.suptitle('Histograms')
ax.set_title('Two Frequency Distributions')

plt.show()
```

```python
# 3
N = 500
normal = np.random.normal(loc=0.0, scale=1.0, size=N)
lognormal = np.random.lognormal(mean=0.0, sigma=1.0, size=N)
index_value = np.random.random_integers(low=0, high=N-1, size=N)
normal_sample = normal[index_value]
lognormal_sample = lognormal[index_value]
box_plot_data = [normal,normal_sample,lognormal,lognormal_sample]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.set_xlabel('Distribution')
ax.set_ylabel('Value')
ax.set_title('Box Plots: Resampling of Two Distributions')
label = ['normal', 'normal_simple', 'lognormal', 'lognormal_sample']
ax.boxplot(box_plot_data, sym='.', notch=False, vert=True, whis=1.5, showmeans=True, labels=label)

plt.show()
```

```python
# 4
import matplotlib as mpl

x1 = np.random.randn(100)
x2 = np.random.randn(100)

plt.style.use('default')
fig, axis = plt.subplots(2, 2, figsize=(6, 6), sharex=True, sharey=True, squeeze=False)

x1 = np.random.randn(100)
x2 = np.random.randn(100)

axis[0, 0].set_title("Uncorrelated")
axis[0, 0].scatter(x1, x2)

axis[0, 1].set_title("Weakly positively correlated")
axis[0, 1].scatter(x1, x1+x2)

axis[1, 0].set_title("Weakly negatively correlated")
axis[1, 0].scatter(x1, x2-x1)

axis[1, 1].set_title("Strongly correlated")
axis[1, 1].scatter(x1, x1 + 0.15 * x2)

axis[1, 1].set_xlabel('x')
axis[1, 0].set_xlabel('x')
axis[0, 0].set_ylabel('y')
axis[0, 1].set_ylabel('y')

plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95)
```

```python
# 5
import seaborn as sns
import pandas as pd

sns.set(color_codes=True)

mean, cov = [5, 10], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
data_frame = pd.DataFrame(data, columns=["x", "y"])
sns.jointplot(x="x", y="y", data=data_frame, kind="reg").set_axis_labels("x", "y")
plt.suptitle("Joint Plot of Two Variables with Bivariate and Univariate Graphs")
plt.show()
```
